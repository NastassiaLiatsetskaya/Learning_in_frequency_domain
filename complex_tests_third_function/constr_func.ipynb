```python
import numpy as np
from numpy.fft import rfft, irfft
def low_fidelity(x):
    return np.sin(8*np.pi*x)
```

```python
x=np.linspace(0,1,400)
x2=np.linspace(0,1,800)
points1=[low_fidelity(i) for i in x ]
points2=[low_fidelity(i) for i in x2 ]

f1_2=[]
points_2=[]
f1=[]
i=0
points=[]
while i< len(points1):
   
    f1.append(points1[i:i+8])
   
    points.append(x[i:i+8])

    i+=8
i=0
while i< len(points2):
   
    f1_2.append(points2[i:i+8])
   
    points_2.append(x2[i:i+8])

    i+=8
```

```python
    
f1_real = []


for i in range(50):
    value1 = rfft(f1[i])
    f1_real.append(irfft((value1)**2+4+1j*5))
    
f1_real=np.asarray(f1_real).flatten()   

f1_2_real = []


for i in range(100):
    value1 = rfft(f1_2[i])
    f1_2_real.append(irfft((value1)**2+4+1j*5))
    
f1_2_real=np.asarray(f1_2_real).flatten()    
import matplotlib.pyplot as plt
plt.plot(x, f1_real, c="b")
plt.plot(x,points1, c="r")
```
